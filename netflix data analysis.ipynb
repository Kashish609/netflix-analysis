```python
# Keep data from year 2000 onwards
df_recent = data[data['release_year'] >= 2000].copy()

# Split 'listed_in' into separate genres
df_recent['listed_in'] = df_recent['listed_in'].astype(str).str.split(', ')
df_exploded = df_recent.explode('listed_in')

# Find top 10 genres overall
top_10_genre = df_exploded['listed_in'].value_counts().head(10).index

# Filter only top 10 genres
df_top_10_genre = df_exploded[df_exploded['listed_in'].isin(top_10_genre)]

# Count how many titles per year per genre
genre_year_counts = (
    df_top_10_genre.groupby(['release_year', 'listed_in'])
    .size()
    .reset_index(name='count')
)

# Plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=genre_year_counts, x='release_year', y='count', hue='listed_in', marker='o')
plt.title('Top 10 Genres Over the Years', fontsize=14, fontweight='bold')
plt.xlabel('Release Year', fontsize=12)
plt.ylabel('Number of Titles', fontsize=12)
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
df_movies  = data[data["type"] == "Movie"].copy()
df_movies['duration_min'] = df_movies['duration'].str.extract('(\d+)').astype(float)
df_movies['main_genre'] = df_movies['listed_in'].str.split(',').str[0]
plt.figure(figsize=(10, 5))
sns.boxplot(data=df_movies, x='main_genre', y='duration_min')
plt.xticks(rotation=90)
plt.title('Movie Duration by Main Genre')
plt.xlabel('Main Genre')
plt.ylabel('Duration (minutes)')
plt.tight_layout()
plt.show()
```
